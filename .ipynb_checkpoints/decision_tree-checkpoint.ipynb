```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('Iris.csv',header=None)
```

```python
df.columns = ["sepal-length","sepal-width","petal-length","petal-width","class"]
```

```python
df.isnull().any()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['sepal-length'].plot.hist()
plt.show()
df['sepal-width'].plot.hist()
plt.show()
df['petal-length'].plot.hist()
plt.show()
df['petal-width'].plot.hist()
plt.show()
```

```python
seaborn.pairplot(df, hue='class')
plt.show()
```

```python
all_inputs = df[['sepal-length', 'sepal-width', 'petal-length', 'petal-width']].values
all_classes = df['class'].values
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes,shuffle=True, train_size=0.7, test_size=0.3,random_state=1)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_inputs, train_classes)
dtc.score(test_inputs, test_classes)
```
